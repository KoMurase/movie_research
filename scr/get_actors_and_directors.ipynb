```python
import pandas as pd 
import numpy as np 

genre = 'SF'

path = r'C:\Users\mkou0\Desktop\movie_research\csv'

data = pd.read_csv(path + '\{}.csv'.format(genre)) #,use_columns=[])
```

```python
data
```

```python
directors = []
target = "監督"

for i in data[data[target] != '[]'][target].values:
    people = i.replace('"', '').replace('[','').replace(']','').replace('\'','')
    #print(people)
    if ',' in people:
        for j in people.split(', '):
            if j not in directors:
                #print(j)
                directors.append(j)
            
    elif people not in directors:
        directors.append(people)
        
 
```

```python
directors
```

```python
for people in directors:
    if '・' in people:
        people = people.replace('・','')
        directors.append(people)
```

```python
directors[-30:-1]
```

```python
writers = []
target = "脚本"
for i in data[data[target] != '[]'][target].values:
    people = i.replace('"', '').replace('[','').replace(']','').replace('\'','')
    #print(people)
    if ',' in people:
        for j in people.split(', '):
            if j not in writers:
                #print(j)
                writers.append(j)
    elif people not in writers:
        writers.append(people)
        
```

```python
writers[-1]
```

```python
casts = []
target = "キャスト"

for i in data[data[target] != '[]'][target].values:
    people = i.replace('"', '').replace('[','').replace(']','').replace('\'','')
    #print(people)
    if ',' in people:
        for j in people.split(', '):
            if j not in casts:
                #print(j)
                casts.append(j)
    elif people not in casts:
        casts.append(people)
        
 
```

```python
save_path = r'C:\Users\mkou0\Desktop\movie_research\エンドロールに出てくる人たち'
pd.to_pickle(directors, save_path + '\directors.pkl')
pd.to_pickle(writers, save_path + '\writers.pkl')
pd.to_pickle(casts, save_path + '\casts.pkl')
```
